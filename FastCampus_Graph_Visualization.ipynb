#<strong> Graph Visualization</strong>

##<strong> Load Package</strong>

```python
import numpy as np
import matplotlib.pyplot as plt
```

##<strong> Basic Atribute</strong>
- alpha : 투명도
- kind : 그래프 종류 : 'line','bar','barh','kde'
- logy : Y축에 대해 Log scaling
- use_index : 객체의 색인을 눈금 이름으로 사용할 지 여부
- rot : 눈금 이름 돌리기 (rotating) 0 ~ 360
- xticks, yticks : x, y 축값으로 사용할 값
- xlim, ylim : X,Y축의 한계값
- grid : 축의 그리드를 표현할 지 여부

- subplots  : 각 columns에 독립된 subplot 그리기
- sharex, sharey : subplots = True 이면 같은 X,Y축을 공유하고 눈금과 한계를 연결
- figsize : 생성될 그래프의 크기를 tuple로 지정
- title  :그래프의 제목 지정
- legend : subplot의 범례지정
- sort_columns : columns을 알파벳 순서로 그린다.


##<strong> Matplotlib </strong>


- 점선 그리기


```python
data = np.random.randn(50).cumsum()
data
```

```python
plt.plot(data)
plt.show()
```

- 여러 그래프 그릴 준비 하기


```python
plt.subplot(1,2,1)# (행,렬,첫번째 것)
plt.subplot(1,2,2)
plt.show()
```

- MultiGraph 그리기


```python
plt.plot(data)
```

```python
hist_data = np.random.randn(100)
scat_data = np.arange(30)
```

```python
plt.subplot(2,2,1)
plt.plot(data)
plt.subplot(2,2,2)
plt.hist(hist_data)
plt.subplot(2,2,3)
plt.scatter(scat_data, np.arange(30) * np.random.randn(30))
plt.show()
```

- 그래프 선 옵션


```python
plt.plot(data)
plt.show()
```

```python
plt.plot(data,'g:')
plt.show()
```

```python
plt.plot(data,'bv')
plt.show()
```

- 그래프 사이즈조절
  - 사이즈는 제일 상단에 위치 해야함
  

```python
plt.figure(figsize=[5,5])
plt.plot(data,'k+')
plt.show()
```

```python
plt.figure(figsize=[10,10])
plt.subplot(2,2,1)
plt.plot(data)
plt.subplot(2,2,2)
plt.hist(hist_data)
plt.subplot(2,2,3)
plt.scatter(scat_data, np.arange(30) * np.random.randn(30))
plt.show()
```

- 그래프 겹치기 + legend 달기


```python
data = np.random.randn(30).cumsum()
plt.plot(data,'b--',label='Default')
plt.plot(data,'k-', drawstyle='steps-post',label='steps-post')
plt.legend()
plt.show()
```

- 이름달기


```python
data = np.random.randn(50).cumsum()
plt.title('Random Graph')
plt.plot(data,'b--',label='Default')
plt.plot(data,'k-', drawstyle='steps-post',label='steps-post')
plt.legend()
plt.show()
```
