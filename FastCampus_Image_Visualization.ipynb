##<strong>이미지 시각화</strong>

###<strong>Package 불러오기</strong>

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

###<strong>이미지 불러오기</strong>

```python
path = '/content/drive/MyDrive/Data_Folder/data/ImageData/dog.jpg'
image_pil = Image.open(path)
image = np.array(image_pil)
```

###<strong>이미지 들여다보기</strong>

```python
image.shape #(가로, 세로, 색상(RGB))
```

```python
np.min(image), np.max(image)
```

###<strong>그래프로 시각화 하기</strong>


```python
plt.hist(image.ravel(),256,[0,256])
plt.show()
```

###<strong>그림 나타내기</strong>

```python
plt.imshow(image)
plt.show()
```

###<strong>이미지 흑백으로 열기</strong>

```python
image_pil = Image.open(path).convert('L')
image_bw = np.array(image_pil)
```

```python
image_bw.shape #흑백으로 변해서 뒤의 컬러값이 없음
```

###<strong>흑백 이미지 열기</strong>

```python
plt.imshow(image_bw,'gray')
plt.show()
```

##<strong>다른색상으로 cmap 표현하기</strong>


###<strong>GrayScale</strong>

```python
plt.imshow(image_bw,'gray')
plt.show()
```

###<strong>RdBu</strong>

```python
plt.imshow(image_bw,'RdBu')
plt.show()
```

###<strong>jet</strong>

```python
plt.imshow(image_bw,'jet')
plt.show()
```

###<strong>Colorbar 추가하기</strong>

```python
plt.imshow(image_bw,'jet')
plt.colorbar()
plt.show()
```

##<strong>이미지 설정</strong>


###<strong>이미지 보기 사이즈 조정</strong>


```python
plt.figure(figsize=[10,10])
plt.imshow(image_bw,'jet')
plt.colorbar()
plt.show()
```

###<strong>이미지에 제목 추가</strong>


```python
plt.title('dog')
plt.imshow(image_bw,'jet')
plt.show()
```
