#<strong>확률</strong>

- Random experiment : 다음과 같은 속성을 지닌 관찰이나 인위적인 실험
  - 실험의 결과는 미리 알 수 없다.
  - 실험에서 일어 날 수 있는 모든 결과는 사전에 알려져 있다.
  - 이론적으로는 실험을 반복 할 수 있다.
- Sample Space(표본공간) : 모든 결과들의 모임(근원사건의 집합)
- Sample outcome(근원사건) : 표본 공간의 원소
- Event(사건) : 표본공간의 부분집합, 근원사건의 집합
  - Mutually exclusive events : 서로 교집합이 공집합인 사건


image.png

- RandomExperience : 주사위를 던지는 시행, 주사위 눈의 숫자로 결과를 표시
- SampleSpace : S = {1,2,3,4,5,6}
- SampleOutcome : 1,...6
- Event : 짝수가 나오는 사건 A = {2,4,6}

- RandomExperience : 두 동전을 던지는 시행 H,T의 쌍으로 결과를 표시
- SampleSpace : S={(H,H),(H,T),(T,H),(T,T)}
- SampleOutcome : (H,H),(H,T),(T,H),(T,T)
- Event : 앞면이 한번이라도 나오는 사건 A = {(H,H),(H,T),(T,H)}


- 어떠한 사건이 일어날 가능성의 정도
  - P({2,4,6}) = P(A)로 표기
- 근원사건이 일어날 가능성이 동일할 때의 계산


image.png

확률의 공리
 - 0 <= P(A) <= 1
 - P(S) = 1
 - 어떠한 사건들이 서로 배반사건일 때, 이사건들의 합 사건의 확률은 각각의 사건이 일어날확률의 합과 같음
 

image.png

- 조건부 확률
  - 사건 B에 대한 정보가 주어졌을때 사건 A의 교정 된 확률
  - B가 주어졌을 때 사건 A의 조건부 확률



image.png

image.png

- 독립
  - 사건 A와 B가 서로에게 아무런 영향을 미치지 않을 때

image.png

- 확률변수
  - 각각의 근원사건들에 실수값을 대응시키는 함수
  - 예) 두 쌍의 동전을 던지는 확률 실험에서 X: 동전의앞면 개수
  - X((H,H)) = 2
  - X((H,T)) = 1
  - X((T,H)) = 1
  - X((T,T)) = 0

- 확률분포
  - 확률변수에서 확률값으로의 함수, 주로 f(x)로 표기
  - f(2) = P(X=2) = P({(H,H)}) = 1/4
  - f(1) = P(X=1) = P({(T,H),(H,T)}) = 1/2
  - f(0) = P(X=0) = P({(T,T)}) = 1/4

#<strong>이산형 확률분포</strong>

- 베르누이 시행
  - 실험의 결과의 범주가 2가지인 경우(성공/실패)
  - X=1(성공) / X=0(실패)

image.png

- 이항분포
  - 성공확률이 p인 베르누이 시행을 독립적으로 n번 시행했을 때 성공한 횟수의 분포
  

image.png

- 다항분포
  - 다항시행 : 1회의 시행결과로 나올 수 있는 범주가 3개이상이 되는 확률 시험
  - K개의 다항 시행을 N번 반복했을 때, 각 범주가 나타나는 횟수의 분포
  

image.png

- 포아송 분포
  - 주어진 단위 구간 내에 **평균적**으로 발생하는 사건의횟수가 정해져 있을 때, **동일단위에서의 발생 횟수**
    - 사건의 평균 발생횟수는 **단위구간**에 비례
    - 두개 이상의 사건이 동시에 발생할 확률은 0에 가깝다
    - 어떤 단위구간의 사건의 발생의 다른 단위구간의 발생으로부터 **독립적**

#**통계적 검정**

- 대립가설 (H1)
  - 입증하여 주장하고자 하는 가설
- 귀무가설 (H0)
  - 대립가설의 반대가설
  - 귀무가설이 아니라는 충분한 증거를 데이터로부터 보임으로써 대립가설을 입증
  - 귀무가설 하에서 통계량의 분포를 아는것이 검정의 핵심
  - 값에서 많이 떨어질수록 귀무가설이 아닐 가능성이 높다

#**검정통계량과 관련된 분포**

- Z 통계량
  - 표본 정규분포로 가게된다
  - 실질적으로 분모의 값을 모름(모집단의 분산)
    - 이값을 S^2이라는 표본분산으로 대체 해야함
    - 관측치의 통계량이 충분하다면, z분포로 근사한다고 할 수 있다.

image.png

- t분포
  - 관측치의 수가 충분하지 않을 때 사용한다.
    - (n-1)이 t분포가 되게 된다.
    - 이 자유도가 커질수록 n-1의 자유도를 가지는데(관측치N개)
    - n이 커질수록 정규분포에 근사하게 된다.

image.png

image.png

- 카이제곱 분포
  - 자유도가 1인 카이스퀘어를 곱해주게되면 자유도가 k개인 카이제곱 분포가 되게된다.
  - 정규분포를 제곱하면 카이스퀘어를 나오고 여기서 k개를 곱해주면 자유도가 k개인 카이제곱분포가 되게 된다.

  - 회귀분석 제곱합 -> 비율을 취할 때 F분포를 따르게 되기 때문에 사용 한다.
