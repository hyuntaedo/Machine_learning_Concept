# <strong>선형 회귀란?</strong>
- 머신러닝의 가장 큰 목적은 실제 데이터를 바탕으로 모델을 생성해, 다른 입력값을 넣었을 경우 발생할 아웃풋을 예측하는데에 있다.

- 이때 우리가 가장 직관적으로 찾아낼 수 있는 방법은 선(line)이다. 그래서 데이터를 놓고 그것을 가장 잘 설명할 수 있는 선을 분석하는 방법을 선형회귀(Linear Regression)이라 한다.

- 예를들어 키와 몸무게의 데이터를 놓고 그것을 잘 설명할 수 있는 선을 하나 그어놓게 된다면 특정인의 키를 바탕으로 몸무게를 예측할 수 있다.


image.png

<strong>y = mx+b</strong>
- 기울기 m, 절편b에 따라 선의 모양이 정해지기 때문에 x를 넣었을 때 y를 구할 수 있다.
- 선형 회귀분석의 최종 목적도 결국 잘 설명할 수 있는 m과 b를 얻는 것이다.



#<strong>선형회귀에서 발생하는 오차 및 손실</strong>
- 데이터를 놓고 선을 긋는다는 것은, 결국 어림잡아서 본다는 뜻인데, 당연히 실제 데이터와 선은 약간의 치이가 발생하게된다. 이를 일종의 오차라 할 수 있는데 이 값이 Loss값이다.


image.png

- 하지만 엄밀히 보자면 ' + '나 ' - ' 방향을 고려하지 않고 설명 한 것이다. 선과 실제 데이터 사이에 얼마나 오차가 있는지 구하려면 양수, 음수에 관계없이 동일하게 반영되도록 모든 손실에 제곱을 구해주는 것이 좋다.
- 그리고 이런 방식으로 손실을 구하는 것을 평균제곱오차(mean squared error, MSE)라고 부른다.

- 손실을 구하는 이외의 방법으로는 MSE처럼 제곱으로 하지말고, 그냥 절대값으로만 바로 평균을 구하는 평균절대오차(MAE), MSE와 MAE를 절충한 후버손실(Huber loss) '1-MSE/VAR'으로 구하는 결정계수(coefficient of determination)이 있다.


- 결국 선형회귀 모델의 목표는 <strong>모든 데이터로부터 나타나는 오차의 평균을 최소화 할수 있는 최적의 기울기와 절편을 찾는것</strong>이다.

#<strong>학습률(Learning Rate)</strong>

image.png

- 학습률이란 것을 정해줄 필요가 있다.
 - 학습률이 너무 높다면 최적의 값을 제대로 찾지 못한다.
 - 하지만 반면에 학습률을 적게 설정한다면 최적의 값을 수렴할 때 까지 시간이 너무 오래 걸리게 된다
- 따라서 모델을 학습 시킬 때는 최적의 학습률을 찾는것이 중요하다.
