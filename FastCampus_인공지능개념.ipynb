## <strong>Data</strong>
- 학습시키기 위한 데이터, 이 데이터가 모델에 들어감
- 데이터가 생성되고, 데이터에 transform 변형을 준다거나, 모델에 들어가기전에 전처리를 한다.
- 이때 들어갈 때 Batch로 만들어서 Model에 넣어줌


##Model
- LeNet, AlexNet, VGG, ResNet등 다양하게 설계된 모델
- Convolution Layer, Pooling등 다양한 Layer층들으로 구성
- 이모델안에 학습 파라미터가 있고, 이 모델이 학습하는 대상


## Loss / Cost
- 예측한 값과 정답과 비교해서 얼마나 틀렸는지를 확인
- Corss Entropy등 다양한 Loss function들이 있음
- 이 때 계산을 통해 나오는 값이 Loss(Cost, Cost Value)이라고 불림
- 이 Loss는 '얼마나 틀렸는지'를 줄이는것

## Optimization
- 앞에서 얻는 Loss값을 최소화 하기 위해 기울기를 받아 최적화된 Varibale값들로 반환
- 이 반환된 값이 적용된 모델은 바로 전에 돌렸을 때의 결과보다 나아지게 됨
- 이 때 최적화된 값만큼 바로 움직이는 것이 아닌 Learning Rate만큼 움직인 값이 적용

## Result
- 평가할 때 또는 예측된 결과를 확인할 때는 예측된 값에서 argmax를 통해 가장 높은 값만을 예측한 class라고 해 둠
# <strong>[0.15, 0.3, 0.2, 0.25, 0.1]</strong>
- 위의 예측값에서 0.2가 제일 높은 값이므로 클래스 2가 가장 높다고 봄(파이썬에서는 0부터 시작)

## <strong>딥러닝 용어</strong>
- Model
  - 딥러닝을 하면서 CNN, VGG, ResNet등이 모델임
  - 이 모델을 학습시키기위해서 다른 것(function)들이 나옴

- Layer
  - 층을 여러층을 쌓은것, Layer는 딥러닝의 핵심
  - Input Layer - Hidden Layer - Output Layer
  - 어떻게 쌓느냐가 모델의 구조가 된다.
  - 몇층을 쌓았느냐에 따라서 성능이 달라짐(층수가 깊으면 좋음 -> 대신 너무 깊다면 Overfitting이 되어 느릴 수 있음)

- Convolution
  - Input Image -> Convolution Keranl(합성곱) -> Feature Map
  - 이미지를 넣었을 때 필터(고정된 값이 아님)를 해서 결과값이 나오게 됨

- Weight / Filter / Keranl / Variable / Bias
  - Weight : 학습하려는 대상
  - y = wx + b

- Pooling Layer
  - Feature를 뽑아서 압축을 하는 것
  - 64 X 64 라면 32 X 32 형식으로 사이즈를 줄여 줌

- Activation Function
  - 앞에서 특징을 뽑아서 음수나 필요 없는 값을 버림(ReLu, Maxout, ELU Sigmoid 등)
  - 불필요 데이터 제거

- Softmax
  - 마지막에 예측을할 때, 각 클래스 마다 전체에서 제일 높은 값(정확한 값)을 유도 하는것 
  - 합쳐서 1이 됨

- Optimization 
  - 최적화
  - 어떻게 하면 새로 업데이트 된 모델이 Loss를 줄일 수 있는가 하는것
  - Loss를 줄이기 위해서 하는 역할

- Learning Rate
  - 기울기가 너무 낮다면 성능은 좋지만 속도가 너무 느릴 수 있다.
  - 사람들이 조정함

- Batch Size
  - 모델에 데이터를 집어 넣는데 모든 데이터(MNIST : 6만개)를 한 번에 넣을 수 없음
  - 몇장을 나누어서 조금 씩 넣어줌
  - 몇장을 넣어 줄 것을 정해주는 것이 Batch Size이다.

- Epoch / Step
  - Epocb 수 만큼 여러번 반복하는것
  - 전체데이터 수를 한 번 볼때 1 Epoch 두 번 볼 때 2 Epoch
- Train / Validation / Test
  - Dataset을 받으면 Trainset과 Testset으로 나눔
  - train과 test로 나누는 이유는 학습을 잘 하기 위해서
  - train으로 학습시키고 test로 결과를 나타냄
- Label / Ground Truth
  - 정답을 Label Ground Truth라고 부름
  - 이게 없다면 Unsupervised Learning이라고 부름
  - 있다면 supervised Learning이다.
  - 모든 데이터 셋에는 정답이 있어야 한다.
