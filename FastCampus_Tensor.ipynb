##<strong>1차원</strong>

```python
arr = np.array([1,2,3,4,5])
print(arr.shape)
print(arr.ndim)
```

- 명심해야 할 것으로는 위의 (3,)은 3이라는 값이 들어간 것이 아닌 shape라는 것을 명심
- 1차원에 3개의 값이 들어갔다는 의미
- 해석하자만 1차원에 3개의 value가 들어가 있다는 것이다.


```python
arr_2 = np.array([1,2,3])
print(arr_2.shape)
print(arr_2.ndim)
```

##<string>2차원</string>

- 대괄호를 추가적으로 씌우면 차원이 추가로 하나 더 생김
- 0차원의 숫자에 대문자 대괄호를 2번씌우면 두개의 차원이 된다.

```python
arr = np.array([[1,2,3],[4,5,6]])
print(arr.shape)
print(arr.ndim)
```

##<strong>Broadcast</strong>
- tensorflow나 pytorch로 계산하면 broadcast를 잘 이해하고 있어야 한다.
- boradcast는 연산하려는 서로 다른 두개 행렬의 shape가 같지않고 한쪽의 차원이라도 같거나 또는 값의 갯수가 한개일 때, 이를 여러가지로 복사를 하여 연산을 함


```python
arr= np.arange(1,10).reshape(3,3)
arr
```

```python
arr + 3
```

```python
arr * 3 
```

```python
arr + np.array([1,1,1])
```
