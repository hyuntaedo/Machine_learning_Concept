 # **모델의 분류와 정답**
 

image.png

- True Positive(TP) : 실제 True인 정답을 True라고 예측(정답)
- False Positive(FP) : 실제 False인 오답을 True라고 예측(오답)
- False Negative(FN) : 실제 True인 정답을 False라고 예측(오답)
- True Negative(TN) : 실제 False인 오답을 False라고 예측(정답)

# **Precision(정밀도)**

- 정밀도랑 모델이 True라고 분류한 것 중에 실제 True인 것의 비율


image.png

- Positive(정답률), PPV(Positive Predictive value)라고도 불린다. 날씨 예측모델이 맑다고 예측했는데 실제로 맑았는지를 살표보는 지표라고 할 수있다.


# **Recall(재현율)**


- 재현율이란 실제 True인 것중 모델이 True라고 예측한것의 비율


image.png

- 통계학에서는 Sensitivity, 그리고 다른 분야에서는 hit rate라는 용어로도 사용한다.
- 실제 날씨가 맑은 날 중에서 모델이 맑다고 예측한 비율을 나타낸 지표, 정밀도와 True Positive의 경우를 다르게 바라보는 것
- 즉 Precision이나 Rcall은 모두 실제 True인 모델이 True라고 예측한 경우에 관심이 있으나, 바라보고자 하는 관점만 다른 것이다.
- Precision은 모델의 입장에서, 그리고 Recall은 실제 정답(data)의 입장에서 정답을 정답이라고 맞춘 경우를 바라보고 있음


**"어떤 요소에 의해 확실히 맑은 날을 예측할 수 있다면 해당하는 날에만 맑은날이라 예측하면 되겠다."**


- 이 경우에는 아예 예측을 하지 않고 보류하여, FP의 수를 줄여서, Precision을 극도로 끌어올리는 편법중 하나이다.
- 예를들어 한달 30일 동안 맑은날이 20일이 있다고 예측한다면, 당연히 맑다고 한 날중에서 실제로 맑은 날은 100%가 나오게 된다. 하지만 과연 이러한 모델이 이상적인 모델인 것인가?


- 우리는 실제 맑은 20일 중에서 예측한 맑은 날의 수도 고려해야한다. 이 경우는 Precision만큼 높은 결과가 나오지는 않지만 Precison과 Recall을 함꼐 고려하면 실제맑은 날의 입장에서 나의 모델이 맑다고 예측한 비율과 함꼐 고려하게되어 제대로 평가가 가능하게 된다.


## **Confusion_Matrix**

image.png

- 서로 Trade-off관계에 있음
  - Type 1 error : P(reject H0|H0 is true)
  - Type 2 error : P(accept H0|H0 is not true)



image.png

- 가설 검정 시에 어떤 상황에서 가설을 받아들일지의 기준이 필요함
- 그 기준으로써 critical region을 잡게 되는데 Type 1 error의 경우, H0가 true일 때, reject H0일 확률, 즉 미리 설정해둔 critical region의 표본을 뽑을 확률이다.
위 그림에서 Any mean이 기준점이고 H0의 관점에서 빨간색 영역이 기각역이다.
이 기각역에 따라 Type 2 error가 정해지고, Any mena을 조정 할 시, Type 1,2 error의 크기가 변하게 된다.

- 하지만 둘 다 커지거나 작아지는 경우가 없는 trade-off관계이다.


- 다시 본론으로 돌아와서, Precision과 Recall은 TP를 분자로써 같이하고 분모에는 TP에 Type 1, 2 error에 해당하는 FN, FP를 더하여 계산. 이때, FN, FP는 각각 Type 1, 2 error에 있으므로 Precision과 Recall 또한 trade-off 관계에 있다고 할 수 있다


# **Accuracy(정확도)**


- True를 True라고 정확하게 예측한 것과 False를 False라고 정확하게 예측한 경우에 대해서 이를 고려하는 지표가 Accuracy라고 한다.


image.png

- 정확도는 가장 직관적으로 모델의 성능을 나타낼 수 있는 지표, 하지만 고려해야하는것이 바로 domain의 편중(bias)이다.
- 만약 우리가 예측하고자 하는 한달동안이 특정 기후에 부합하여 비오는날이 흔치 않다고 생각한다면, 이 경우는 data의 domain이 불균형하게 되므로 맑은 것을 예측하는 성능은 높지만 비가 오는 것을 예측하는 성능은 낮을 수 밖에 없다.
따라서 보완할 지표가 필요하다


# **F1-Score**


- F1-Score는 label이 불균형 할 때, 모델의 성능을 정확하게 평가할 수 있으며, 성능을 하나의 숫자로 표현이 가능하다.
- 여기서 단순 산술평균으로 사용하지 않는 이유는, 조화평균이 낫기 떄문이다.
