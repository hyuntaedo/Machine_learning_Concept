#<strong>두점사이의 거리공식(Distance Formula)</strong>

- <strong>거리를 왜 구해야하나?</strong> 거리는 일종의 <strong>유사도 (Similarity)</strong>개념이기 때문이다.
- 즉 거리가 가까울 수록 그 특성(feature)들이 비슷하다는 뜻이다.
- 그러므로 머신러닝 알고리즘에서도 매우 널리 유용하게 사용된다.


##<strong>두 점사이의 거리를 구하는 방법 3가지</strong>
- 유클리드 거리(Euclidean Distance)
- 맨하탄 거리(Manhattan Distance)
- 해밍 거리(Hamming Distance)

#<strong>1. 유클리드 거리(Euclidean Distance)</strong>
- 유클리디안 거리라고 영어단어를 그대로 읽기도 한다.
- 예를 들어 아래와 같이 2차원 점에 있는 점 a와 점 b의 거리를 구한다면 이렇게 나타낼 수 있다.


image.png

- 피타고라스의 정리다.
- 위의 예제는 2차원이지만 만약 n차원에서 두 점사이의 거리를 구한다면 

image.png

이런식으로 나타 낼 수 있다.


```python
def euclidean_distance(pt1,pt2):
    distance =0;
    for i in range(len(pt1)):
        distance += (pt1[i] - pt2[i]) ** 2
    return distance ** 0.5
```

```python
euclidean_distance([3,4],[1,2])
```

이런식으로 결과 값이 나오게 된다.


#<strong>2. 맨하탄 거리(Manhattan Distance)</strong>
- 맨하탄 거리도 유클리드 거리와 유사하긴 하지만 각 차원의 차를 제곱해서 사용하는것이 아닌, 절대값을 바로 합산하는 것이다.
- 아래 그림같은 개념으로 이해하면 쉽게 이해가 가능하다.

image.png

- 그림에서 알 수 있듯이 <strong>맨하탄 거리는 항상 유클리드 거리보다 크거나 같다.</strong>
- 만약 n차원에서 두 점사이의 거리를 구한다면 이렇게 나타낼 수 있다.


image.png

```python
def manhattan_distance(pt1,pt2):
    distance = 0
    for i in range(len(pt1)):
        distance += abs(pt1[i] - pt2[i])
    return distance
```

```python
manhattan_distance([3,4],[10,6])
```

이런식으로 결과값이 나오게 된다.


#<strong>3. 해밍 거리(Hamming Distance)</strong>
- 해밍거리는 앞서 설명한 유클리드, 맨하탄 거리와는 좀 다른 개념이다.
- 해밍거리는 각 차원마다 차이를 찾는것이 아닌, '정확히 같은지' 여부만 고려한다.
- 그러므로 주로 맞춤법검사 같은 알고리즘에 자주 사용된다.


```python
def hamming_distance(pt1,pt2):
    distance = 0
    for i in range(len(pt1)):
        if pt1[i] != pt2[i]:
            distance +=1
    return distance
```

```python
hamming_distance([5,4,3],[1,9,7])
```

#<strong>Scipy를 통해 거리 구하기</strong>
- 이 3가지의 거리 구하는 방법을 Scipy를 통해서 간단하게 구할 수 있다.
  - 유클리드거리 `.euclidean()`
  - 맨하탄거리 `.cityblock()`
  - 해밍거리 `.hamming()`
- 단 해밍거리는 Scipy에서는 0과 1사이의 값으로 돌려준다. 그래서 마지막 hamming의 값은 2/3이 되는 것이다.


```python
from scipy.spatial import distance

print(distance.euclidean([1, 2], [4, 0]))
print(distance.cityblock([1, 2], [4, 0]))
print(distance.hamming([5, 4, 9], [1, 7, 9]))
```
